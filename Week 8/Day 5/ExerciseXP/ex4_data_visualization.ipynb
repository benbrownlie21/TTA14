```python
import pandas as pd

file = (r'C:\Users\bbrow\Desktop\Courses\Developers Institute\Exercises\Week 8\Day 5\ExerciseXP\sales_data.csv')
sales_data = pd.read_csv(file)
sales_data.head(5)
```

```python
total_products_sold = sales_data['quantity'].sum()
print(total_products_sold)
```

```python
grouped_categories = sales_data.groupby('category').sum('revenue')
sorted_categories = grouped_categories.sort_values(by="revenue", ascending=False)
sorted_categories.head()
```

```python
grouped_sales = sales_data.groupby('date').mean('revenue')
grouped_sales = grouped_sales.rename(columns={'revenue': 'avg_revenue'})
grouped_sales.head()
```

```python
sales_data['date'] = pd.to_datetime(sales_data['date'])

sales_data['quarter'] = sales_data['date'].dt.quarter
sales_data.head()
```

```python
import matplotlib.pyplot as plt

x = sales_data['quarter']
y = sales_data['revenue']

plt.bar(x, y)
plt.xlabel('Quarters')
plt.ylabel('Revenue')
plt.title('Revenue per Quarter')
plt.show()
```
