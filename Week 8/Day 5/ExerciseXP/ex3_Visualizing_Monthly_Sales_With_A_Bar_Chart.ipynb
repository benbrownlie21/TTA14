```python
import pandas as pd

data = [[172, 'Jan'], [174, 'Feb'], [124, 'Mar'], [98, 'Apr'], [102, 'May'], [86, 'Jun'], [95, 'Jul'], [123, 'Aug'], [122, 'Sep'], [123, 'Oct'], [101, 'Nov'], [134, 'Dec']]

sales_df = pd.DataFrame(data, columns=['Sales', 'Month'])
```

```python
import matplotlib.pyplot as plt

x = sales_df['Month']
y = sales_df['Sales']

plt.bar(x, y)
plt.xlabel('Months')
plt.ylabel('Number of Sales')
plt.title('Sales over Months')
plt.show()
```
