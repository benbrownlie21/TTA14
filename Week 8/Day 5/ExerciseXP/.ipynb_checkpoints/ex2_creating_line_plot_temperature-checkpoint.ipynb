```python
import pandas as pd

data = [[72, 'Monday'], [74, 'Tuesday'], [76, 'Wednesday'], [80, 'Thursday'], [82, 'Friday'], [78, 'Saturday'], [75, 'Sunday']]

temperature_df = pd.DataFrame(data, columns=['Temperature', 'Day'])
```

```python
temperature_df.head()
```

```python
import matplotlib.pyplot as plt

x = temperature_df['Day']
y = temperature_df['Temperature']

plt.plot(x, y, marker='o')
plt.xlabel('Day of the Week')
plt.ylabel('Temperature')
plt.title('Temperature over Days of the Week')
plt.show()
```
