```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = [['A', 15, 'Mon'], ['B', 30, 'Mon'], ['C', 45, 'Mon'], 
        ['A', 20, 'Tue'], ['B', 25, 'Tue'], ['C', 35, 'Tue'], 
        ['A', 30, 'Wed'], ['B', 45, 'Wed'], ['C', 40, 'Wed'], 
        ['A', 50, 'Thur'], ['B', 55, 'Thur'], ['C', 60, 'Thur']]
sales_df = pd.DataFrame(data, columns=['Product', 'Values', 'Day'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Day', y='Values', hue='Product', data=sales_df)
plt.xlabel('Day of the Week')
plt.ylabel('Sales Values')
plt.title('Sales over Days of the Week by Product')
plt.legend(title='Product')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = [['Apples', 40], ['Bananas', 30], ['Cherries', 20], ['Dates', 10]]
fruit_df = pd.DataFrame(data, columns = ['Fruit', 'Percentage'])

plt.pie(fruit_df['Percentage'], labels=fruit_df['Fruit'])
plt.show()
```
