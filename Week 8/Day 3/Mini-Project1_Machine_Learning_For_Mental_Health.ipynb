In the code below I am replacing the null values from state with the most frequent entries

```python
from sklearn.impute import SimpleImputer
import numpy as np

imputer = SimpleImputer(missing_values = np.nan, strategy ='most_frequent')

state_data = mentalHealth_df[['state']]

imputer = imputer.fit(state_data)

state_data_imputed = imputer.transform(state_data)

mentalHealth_df['state'] = state_data_imputed.flatten()

print("Imputed Data : \n", mentalHealth_df)
```

```python
mentalHealth_df.isna().sum()
```

```python
mental_health_df = mentalHealth_df.dropna(subset=['self_employed'])
```

In the code below I am replacing the null values from work_interfere with the most frequent entries

```python
imputer = SimpleImputer(missing_values = np.nan, strategy ='most_frequent')

work_interfere_data = mental_health_df[['work_interfere']]

imputer = imputer.fit(work_interfere_data)

work_interfere_data_imputed = imputer.transform(work_interfere_data)

mental_health_df.loc[:,'work_interfere'] = work_interfere_data_imputed.flatten()

print("Imputed Data : \n", mental_health_df)
```

What is the distribution of mental health conditions among different age groups in the tech industry?
