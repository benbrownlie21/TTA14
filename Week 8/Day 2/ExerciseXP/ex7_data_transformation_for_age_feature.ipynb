```python
import pandas as pd

file = (r'C:\Users\bbrow\Desktop\Courses\Developers Institute\Exercises\Week 8\Day 2\ExerciseXP\titanic-dataset\Titanic-Dataset.csv')
titanic_data = pd.read_csv(file)
```

```python
age_bins = [0, 18, 35, 60, 100]
age_labels = ['0-18', '19-35', '36-60', '61+']

titanic_data['Age_Group'] = pd.cut(titanic_data['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
titanic_data.head(5)
```

```python
age_group_count = titanic_data['Age_Group'].value_counts() 
print(age_group_count)
```

```python
one_hot_encoded_data = pd.get_dummies(titanic_data, columns = ['Age_Group'])
print(one_hot_encoded_data)
```

```python
merged_titanic_data = titanic_data.merge(one_hot_encoded_data)
merged_titanic_data.head(5)
```
