```python
import pandas as pd
```

```python
df09 = pd.read_csv("/content/Year 2009-2010.csv", encoding='ISO-8859-1')
df10 = pd.read_csv("/content/Year 2010-2011.csv", encoding='ISO-8859-1')
```

```python
df09.head()
```

```python
# prompt: I needd to anonamize the columns Invoice, Customer Id and Country

import hashlib

def anonymize_column(df, column_name):
  """Anonymizes a column in a pandas DataFrame."""
  df[column_name] = df[column_name].apply(lambda x: hashlib.sha256(str(x).encode()).hexdigest())
  return df

# Anonymize the specified columns in both DataFrames
df09 = anonymize_column(df09, 'Invoice')
df09 = anonymize_column(df09, 'Customer ID')
df09 = anonymize_column(df09, 'Country')

df10 = anonymize_column(df10, 'Invoice')
df10 = anonymize_column(df10, 'Customer ID')
df10 = anonymize_column(df10, 'Country')
```

In the above I have two datasets where the invoice number, the customer id and the country are all anonymized using hashing. This could be useful if the data needs to be shared with externals - so then they are unable to read data that is not needed by them. However, internally it makes the data quite useless - it is impossible to analyse the data when it is hashed - you cannot get insights for specific customers or customer groups to improve the customer experience.
