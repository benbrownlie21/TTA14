```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['age'], data['income'])
plt.scatter(data['age'], data['income'], color='blue', label='Data points')

regression_line = slope * data['age'] + intercept
plt.plot(data['age'], regression_line, color='red', label='Regression line')

plt.xlabel('Age')
plt.ylabel('Income (dollars)')
plt.title('Income vs Age')
plt.legend()

plt.show()
```
