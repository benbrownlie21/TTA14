```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
request_url = urllib.request.urlopen('https://www.bbc.com')

soup = BeautifulSoup(request_url, 'html.parser')
print(soup.prettify())
```

```python
with open('index.html') as file:
    html_content = file.read()

my_soup = BeautifulSoup(html_content, 'html.parser')
print(my_soup.prettify())
```

```python
print(my_soup.html.head.title)
```

```python
my_soup.find_all("p")
```

```python
for a in my_soup.find_all('a', href=True):
    print("Found the URL:", a['href'])
```
