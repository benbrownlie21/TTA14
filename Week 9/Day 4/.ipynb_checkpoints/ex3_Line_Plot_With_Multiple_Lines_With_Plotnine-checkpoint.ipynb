```python
import pandas as pd
import numpy as np

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)

data = {
    'x': x,
    'y1': y1,
    'y2': y2
}
df = pd.DataFrame(data)

df.head()
```

```python
from plotnine import ggplot, aes, geom_line, labs
```

```python
df_long = df.melt(id_vars='x', value_vars=['y1', 'y2'], var_name='Function', value_name='Value')

(
    ggplot(df_long, aes(x='x', y='Value', color='Function'))
    + geom_line()
    + labs(
        title="Comparison of Two Functions",
        x="X",
        y="Value"
    )
)
```
