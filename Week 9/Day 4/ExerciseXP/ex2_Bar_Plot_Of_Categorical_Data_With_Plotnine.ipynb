```python
import pandas as pd
import numpy as np

categories = ['Category A', 'Category B', 'Category C']
data = np.random.choice(categories, size=30)

df = pd.DataFrame(data, columns=['Category'])

print(df)
```

```python
cat_count = df.groupby(['Category']).size()
cat_count
```

```python
from plotnine import ggplot, aes, geom_bar, coord_flip, theme_classic, labs, geom_text, after_stat
```

```python
(
    ggplot(df) + geom_bar(aes(x="Category"))
    + labs(title="Category Frequency", x="Category Distribution", y='Category Count')
)
```
