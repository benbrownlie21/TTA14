```python
import pandas as pd

data = {
    'X': [1.1, 2.0, 3.4, 4.6, 5.0, 6.2, 7.1, 8.3, 9.7, 10.2],
    'Y': [2.3, 3.7, 1.9, 4.1, 5.3, 3.8, 5.9, 4.5, 6.1, 5.6]
}

df = pd.DataFrame(data)
df
```

```python
import numpy as np
import pandas as pd
from plotnine import (
    ggplot,
    aes,
    geom_point,
    theme_matplotlib,
    theme_set,
    labs,
)
```

```python
(
    ggplot(df, aes("X", "Y"))
    + geom_point()
    + labs(title="Basic Scatter Plot", x="X-Value", y='Y-Value')
)
```
